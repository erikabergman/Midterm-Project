```python
full_hour['dep_hour'] = full_hour['dep_time'].apply(lambda x: str(x)[:2] if len(str(x)) == 6 else (str(x)[0] if len(str(x)) == 5 else 0))
full_hour['dep_hour'] = full_hour['dep_hour'].replace('24', '0')
full_hour['dep_hour'] = full_hour['dep_hour'].astype('int')
```

```python
full_hour['arr_hour'] = full_hour['arr_time'].apply(lambda x: str(x)[:2] if len(str(x)) == 6 else (str(x)[0] if len(str(x)) == 5 else 0))
full_hour['arr_hour'] = full_hour['arr_hour'].replace('24', '0')
full_hour['arr_hour'] = full_hour['arr_hour'].astype('int')
```
