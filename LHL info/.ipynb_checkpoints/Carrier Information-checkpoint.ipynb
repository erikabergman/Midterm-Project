```python
baseline_df.columns
```

```python
mapping2 = {'dep_delay':'mean_dep_delay', 'arr_delay':'mean_arr_delay', 'distance':'mean_distance'}
mapping3 = {'dep_delay':'med_dep_delay', 'arr_delay':'med_arr_delay', 'distance':'med_distance'}
```

```python
base_del_mean =baseline_df[['op_unique_carrier', 'dep_delay', 'arr_delay', 'distance']].groupby('op_unique_carrier').mean().rename(mapping2, axis = 1)
```

```python
base_del_med = baseline_df[['op_unique_carrier', 'dep_delay', 'arr_delay', 'distance']].groupby('op_unique_carrier').median().rename(mapping3, axis = 1)
```

```python
base_del_med
```

```python
mapping = {'carrier_delay':'mean_carr', 'weather_delay':'mean_weather', 'nas_delay':'mean_nas', 'security_delay':'mean_sec', 'late_aircraft_delay':'mean_late_air'}
```

```python
base_types_mean = baseline_df[['op_unique_carrier','carrier_delay', 'weather_delay', 'nas_delay', 'security_delay', 'late_aircraft_delay']].groupby('op_unique_carrier').mean().rename(mapping, axis = 1)
```

```python
baseline_df['is_arr_delayed'] = 0
baseline_df['is_arr_delayed'].loc[baseline_df['arr_delay'] >= 15] = 1
```

```python
baseline_percdel = baseline_df[['op_unique_carrier', 'is_arr_delayed']].groupby('op_unique_carrier').mean().rename({'is_arr_delayed':'perc_del'}, axis = 1)
```

```python
merge1 = baseline_percdel.merge(base_del_mean, left_index = True, right_index = True)
```

```python
merge2 = merge1.merge(base_del_med, left_index = True, right_index = True)
```

```python
merge3 = merge2.merge(base_types_mean, left_index = True, right_index = True)
```

```python
merge3.head()
```

```python
fuel_carrier.shape
```

```python
carrier_flight_fuel = merge3.merge(fuel_carrier, how = 'outer', left_index = True, right_index = True).drop('airline_id', axis = 1).fillna(0)
```
